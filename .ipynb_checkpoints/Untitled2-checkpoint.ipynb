```python
def clfreport(true_labels,pred_labels):
    
    reports = pd.DataFrame(classification_report(true_labels, pred_labels, digits=3, output_dict=True)).T
    class_name=["False","True"]

    fig, ax = plt.subplots(1, 2, figsize=(20, 8))  
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    cm = confusion_matrix(true_labels, pred_labels)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax[0])
    ax[0].set_xlabel('Predicted labels', fontsize=18)
    ax[0].set_ylabel('True labels', fontsize=18)
    ax[0].set_title('Confusion Matrix', fontsize=25)
    ax[0].xaxis.set_ticklabels(class_name)
    ax[0].yaxis.set_ticklabels(class_name)

    
    accuracy_false = cm[0, 0] / (cm[0, 0] + cm[0, 1])
    accuracy_true = cm[1, 1] / (cm[1, 0] + cm[1, 1])
    reports.loc['False', 'accuracy'] = accuracy_false
    reports.loc['True', 'accuracy'] = accuracy_true
    sns.heatmap(reports.drop(columns="support", axis=1), cmap='Blues', annot=True, linecolor='white', linewidths=0.5, ax=ax[1])
    ax[1].xaxis.tick_top()
    ax[1].set_title('Classification Report', fontsize=25)
```

```python
def clean_df(df,missing_rate=0.9, missing_rate_rows=0.5):
    
    high_missing_features = df.columns[df.isnull().mean() > missing_rate].tolist()
    
    
    print(high_missing_features)
    df = df.drop(columns=[col for col in set(high_missing_features) if col != '1Yr_Death'],axis=1)
    
    
    row_nan_ratio = df.isnull().mean(axis=1)

    df = df[row_nan_ratio <= missing_rate_rows]
    
    return df
```

```python
def fill_missing_values(df, num_strategy="mean", fill_value=None, n_neighbors=5,cat_strategy="freq"):
    
    df_num = df.drop(['1Yr_Death'],axis=1).select_dtypes(include=['int64', 'float64'])
    df_cat = df.drop(['1Yr_Death'],axis=1).select_dtypes(include=['object', 'bool'])
    df_label = df['1Yr_Death']
    
    if num_strategy == "knn":
        imputer_num = KNNImputer(n_neighbors=n_neighbors)
    else:
        imputer_num = SimpleImputer(strategy=num_strategy, fill_value=fill_value)
    
    if cat_strategy=="freq":
        imputer_cat = SimpleImputer(strategy='most_frequent')
    else :
        imputer_cat = SimpleImputer(strategy="constant", fill_value="missing")
        
        
    df_num_imputed  = pd.DataFrame(imputer_num.fit_transform(df_num), columns=df_num.columns, index=df_num.index)
    df_cat_imputed  = pd.DataFrame(imputer_cat.fit_transform(df_cat), columns=df_cat.columns, index=df_cat.index)

    df_filled = pd.concat([df_num_imputed , df_cat_imputed , df_label], axis=1)
    
    #log
    # df_filled["Total LOS Admitted in Last Year"] = np.log(df_filled["Total LOS Admitted in Last Year"]+1)
    # large_val_cols = list(df_filled.drop(["1Yr_Death"],axis=1).describe().loc["mean"][df_filled.drop(["1Yr_Death"],axis=1).describe().loc["mean"]>50].index)
    #     # print(large_val_cols)
    # for c in large_val_cols:
    #     df_filled[c] = np.log(df_filled[c]+1)
    
    
    return df_filled
    
```

```python
def preprocessing(df):
    
    print(df.shape)
    print("Drop Missing...")
    df = clean_df(df,missing_rate=0.5)
    print(df.shape)
    print("Fill nan...")
    df = fill_missing_values(df,num_strategy='median')
    print(df.shape)
    print("Add col...")
    df = add_preprocessing(df)
    print(df.shape)
    print("Drop col...")
    df = drop_preprocessing(df)
    print(df.shape)

    return df
    
```

```python
def set_preprocessor(df):
    
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    cat_cols = df.select_dtypes(include=['object', 'bool']).columns
    
    num_cols = [col for col in num_cols if col != '1Yr_Death']
    cat_cols = [col for col in cat_cols if col != '1Yr_Death']
    
   
    
    numeric_transformer = Pipeline(steps=[
        ('scaler', StandardScaler())])
    
    categorical_transformer = Pipeline(steps=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'))])

    
    preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)])
    
    return preprocessor
```

```python
def set_pipeline(preprocessor=None,imbalance_plan=None,model=None):
    steps=[]
    if preprocessor is not None:
        steps.append(('preprocessor', preprocessor))
    if imbalance_plan is not None:
        steps.append(('imbalance_plan',imbalance_plan))
    if model is not None:
        steps.append(('model',model))
        
    pipeline=ImbPipeline(steps=steps)
    
    return pipeline
```

```python
def run_result(pipeline,X_train, X_test, y_train, y_test ):
    
    print("Original....")
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy}')
    clfreport(y_test,y_pred)
    
    y_pred_proba = pipeline.predict_proba(X_test)[:, 1]
    
    print("Best acc....")
    best_accuracy = 0
    best_threshold_acc = 0.5
    for threshold in np.linspace(0, 1, 101):
        y_pred_adjusted_acc = (y_pred_proba >= threshold).astype(int)
        current_accuracy = accuracy_score(y_test, y_pred_adjusted_acc)
        if current_accuracy > best_accuracy:
            best_accuracy = current_accuracy
            best_threshold_acc = threshold
    print(best_threshold_acc)
    y_pred = (y_pred_proba >= best_threshold_acc).astype(int)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy}')
    clfreport( y_test,y_pred)  

    print("Best F1....")
    thresholds = np.linspace(0, 1, 101)
    best_f1 = 0
    best_threshold_f1 = 0.5
    for threshold in thresholds:
        y_pred_adjusted = (y_pred_proba >= threshold).astype(int)
        report_adjusted = classification_report(y_test, y_pred_adjusted, output_dict=True, zero_division=0)  
        f1 = report_adjusted['True']['f1-score']
        if f1 > best_f1:
            best_f1 = f1
            best_threshold_f1 = threshold
    print(best_threshold_f1)
    y_pred = (y_pred_proba >= best_threshold_f1).astype(int)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy}')
    clfreport( y_test,y_pred)
    return pipeline
     
```

```python
from pytorch_tabnet.tab_model import TabNetClassifier
import torch
```

```python
pipeline = ImbPipeline(steps=[('preprocessor', preprocessor),
                              ('oversampler', RandomOverSampler(random_state=42))])

X_resampled, y_resampled = pipeline.fit_resample(X_train, y_train)


clf = TabNetClassifier(optimizer_fn=torch.optim.Adam,
                       optimizer_params=dict(lr=1e-2),
                       scheduler_params={"step_size":50, "gamma":0.9},
                       scheduler_fn=torch.optim.lr_scheduler.StepLR,
                       mask_type='sparsemax',
                       device_name='cuda'
                      )
clf.fit(X_resampled, y_resampled, eval_set=[(preprocessor.fit_transform(X_test), y_test.values)], max_epochs=100, patience=20, batch_size=32, virtual_batch_size=8)


predictions = clf.predict(preprocessor.fit_transform(X_test))
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
```
