```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset,TabularPredictor
import numpy as np
```

```python
df = pd.read_csv("data/Training_Set.csv")
test = pd.read_csv("data/Test_Set_NO_LABELS.csv")
print(df.shape)

print(df['1Yr_Death'].value_counts())

df['1Yr_Death_Numeric'] = df['1Yr_Death'].astype(int)

death_stats_by_year = df.groupby('Admission Year')['1Yr_Death_Numeric'].agg(
    Sample_Total='count',  
    Death_Count='sum', 
    Death_Ratio=lambda x: x.mean()  
)

df.drop(['1Yr_Death_Numeric'], axis=1, inplace=True)
print(death_stats_by_year)



missing_ratio = df.isnull().mean() * 100
df= df.drop(columns=missing_ratio[missing_ratio > 60].index)
print(df.shape)


# corr_matrix = df.corr()

# high_corr_var = np.where(corr_matrix > 0.9)
# high_corr_var = [(corr_matrix.index[x], corr_matrix.columns[y]) for x, y in zip(*high_corr_var) if x != y and x < y]

# to_keep = set(corr_matrix.columns)

# for var1, var2 in high_corr_var:
#     if var1 in to_keep and var2 in to_keep:
#         to_keep.remove(var2)

# filtered_df = df[to_keep]


# large_val_cols = list(df.describe().loc["mean"][df.describe().loc["mean"]>50].index)
# for c in large_val_cols:
#     df[c] = np.log(df[c]+1)
    

    
    
train_df,test_df = train_test_split(df, test_size=0.3,stratify=df["1Yr_Death"],random_state=42)
```

```python
train_df
```

```python
train_data = TabularDataset(train_df)
to_drop=["Patient_ID","Primary Insurance"] #,"Admission Year"
label = "1Yr_Death"
predictor = TabularPredictor(label = label).fit(train_data.drop(columns=to_drop))

test_data = TabularDataset(test_df)
preds = predictor.predict(test_data.drop(columns=to_drop))

accuracy = accuracy_score(test_df[label], preds)
print(accuracy)
```
